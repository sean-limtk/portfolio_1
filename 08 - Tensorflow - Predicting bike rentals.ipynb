# Using Tensorflow to predict bike sharing.Â¶

Similar to the last project, this notebook attempts to predict bike rides from data set for the bike sharing scheme in Washington, D.C. and the data is compiled by [Capital Bikeshare](https://www.capitalbikeshare.com/system-data).
 
The predictive approach here is the neural network TensorFlow, developed by Google. 

## Imports

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

import sys
```

## Load the data

```python
rides = pd.read_csv('bike_rental_hour.csv')
rides.head()
```

### Dummy variables

```python
dummy_fields = ['season', 'weathersit', 'mnth', 'hr', 'weekday']
for each in dummy_fields:
    dummies = pd.get_dummies(rides[each], prefix=each, drop_first=False)
    rides = pd.concat([rides, dummies], axis=1)

fields_to_drop = ['instant', 'dteday', 'season', 'weathersit', 
                  'weekday', 'atemp', 'mnth', 'workingday', 'hr']
data = rides.drop(fields_to_drop, axis=1)
data.head()
```

### Scaling target variables

```python
quant_features = ['casual', 'registered', 'cnt', 'temp', 'hum', 'windspeed']
# Store scalings in a dictionary so we can convert back later
scaled_features = {}
for each in quant_features:
    mean, std = data[each].mean(), data[each].std()
    scaled_features[each] = [mean, std]
    data.loc[:, each] = (data[each] - mean)/std
```

```python
data.head()
```

### Splitting the data into training, testing, and validation sets

```python
# Save data for approximately the last 21 days 
test_data = data[-21*24:]

# Now remove the test data from the data set 
data = data[:-21*24]

# Separate the data into features and targets
target_fields = ['cnt', 'casual', 'registered']
features, targets = data.drop(target_fields, axis=1), data[target_fields]
test_features, test_targets = test_data.drop(target_fields, axis=1), test_data[target_fields]
```

```python
# Hold out the last 60 days or so of the remaining data as a validation set
train_features, train_targets = features[:-60*24], targets[:-60*24]
val_features, val_targets = features[-60*24:], targets[-60*24:]
```

```python
# for training 
x_train = np.array(train_features)
y_train = np.array([[i] for i in train_targets['cnt']])

# for validation
x_cv = np.array(test_features)
y_cv = np.array([[i] for i in test_targets['cnt']])
```

```python
# Parameters
learning_rate = 0.001
training_epochs = 500
batch_size = 64
display_step = 1
dropout_rate = 0.7


# Network Parameters
n_hidden_1 = 50 # 1st layer number of features
n_hidden_2 = 25 # 2nd layer number of features


total_len = x_train.shape[0]
n_input = x_train.shape[1]

n_classes = 1

tf.reset_default_graph()

# tf Graph input
x = tf.placeholder(tf.float32, [None, 56])
y = tf.placeholder(tf.float32, [None, 1])

keep_prob = tf.placeholder(tf.float32)

# Create model
def NeuralNetwork(x, weights, biases):
    # Hidden layer with RELU activation
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1 = tf.sigmoid(layer_1)
    # Dropout on hidden layer: RELU layer
    hidden_layer_1_dropout = tf.nn.dropout(layer_1, keep_prob)

    # Hidden layer with RELU activation
    layer_2 = tf.add(tf.matmul(layer_1, weights['h2']), biases['b2'])
    layer_2 = tf.sigmoid(layer_2)
    hidden_layer_2_dropout = tf.nn.dropout(layer_2, keep_prob)


    # Output layer with linear activation
    out_layer = tf.matmul(hidden_layer_2_dropout, weights['out']) + biases['out']
    return out_layer

# Store layers weight & bias
weights = {
    'h1': tf.Variable(tf.truncated_normal([n_input, n_hidden_1])),
    'h2': tf.Variable(tf.truncated_normal([n_hidden_1, n_hidden_2])),
    'out': tf.Variable(tf.truncated_normal([n_hidden_2, n_classes]))
}
biases = {
    'b1': tf.Variable(tf.random_normal([n_hidden_1], 0, 0.1)),
    'b2': tf.Variable(tf.random_normal([n_hidden_2], 0, 0.1)),
    'out': tf.Variable(tf.random_normal([n_classes], 0, 0.1))
}

# Construct model
pred = NeuralNetwork(x, weights, biases) 

# Define loss and optimizer
cost = tf.reduce_mean(tf.square(pred-y))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)

init = tf.global_variables_initializer()

losses = {'train':[], 'validation':[]}

# Launch the graph
#with tf.Session() as sess:
sess = tf.Session()
sess.run(init)

# Training cycle
for epoch in range(training_epochs):
    avg_cost = 0.
    avg_val_cost = 0.
    
    total_batch = int(total_len/batch_size)
    # Loop over all batches
    for i in range(total_batch-1):
        batch_x = x_train[i*batch_size:(i+1)*batch_size]
        batch_y = y_train[i*batch_size:(i+1)*batch_size]
        # Run optimization op (backprop) and cost op (to get loss value)
        _, c, p = sess.run([optimizer, cost, pred], feed_dict={x: batch_x, y: batch_y, keep_prob: dropout_rate})
        
        #avg_cost += c / total_batch
    
    avg_cost = sess.run(cost, feed_dict={x: x_train, y: y_train, keep_prob: dropout_rate})   
    avg_val_cost = sess.run(cost, feed_dict={x: x_cv, y: y_cv, keep_prob: dropout_rate})
        
    # Display logs per epoch step
    if epoch % display_step == 0:
        sys.stdout.write("\rProgress: {:2.1f}".format(100 * epoch/float(training_epochs))\
                         + "%... Training loss: " + str(avg_cost)[:5] \
                         + " ... Validation loss: " + str(avg_val_cost)[:5])
                       
        sys.stdout.flush()
        
    losses['train'].append(avg_cost)
    losses['validation'].append(avg_val_cost)
```

```python
plt.plot(losses['train'], label='Training loss')
plt.plot(losses['validation'], label='Validation loss')
plt.ylim([0,1])
plt.legend()
_ = plt.ylim()
```

## Implement Neural Network using Tensorflow

```python
n_dim = x_train.shape[1]


tf.reset_default_graph()


X = tf.placeholder(tf.float32, [None, n_dim])
Y = tf.placeholder(tf.float32, [None, 1])      

W1 = tf.Variable(tf.truncated_normal([n_dim,1]), name="weight")
b1 = tf.Variable(tf.truncated_normal([n_dim]), name="bias")

h1 = tf.nn.relu(tf.matmul(X,W1) + b1)

W2 = tf.Variable(tf.truncated_normal([n_dim,1]), name="weight")
b2 = tf.Variable(tf.truncated_normal([1]), name="bias")


y_ = tf.add(tf.matmul(h1, W2),b2)

cost_function = tf.reduce_mean(tf.square(y_ - Y))
```

### Choose the learning rate

This scales the size of weight updates. If this is too big, the weights tend to explode and the network fails to fit the data. A good choice to start at is 0.1. If the network has problems fitting the data, try reducing the learning rate.

### Choose the number of iterations

In theory, using more iterations would enable the model to better will fit the data
However, using too many iterations could lead to overfitting, causing the model accuracy to be poor with other data.  We will try to find an iteration number that produces low training loss and low validation loss.

```python
# Set hyperparameters
learning_rate = 0.01
training_epochs = 1000
```

```python
# Change optimizer to AdamOptimizer
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost_function)
```

```python
init = tf.global_variables_initializer()

sess = tf.Session()
sess.run(init)

cost_history = np.empty(shape=[1],dtype=float)
cost_val_history = np.empty(shape=[1],dtype=float)

for epoch in range(training_epochs):
    sess.run(optimizer,feed_dict={X:np.array(train_features),Y:y_train})
    cost_history = np.append(cost_history,sess.run(cost_function,feed_dict={X:np.array(train_features),Y:y_train}))
    cost_val_history = np.append(cost_val_history,sess.run(cost_function,feed_dict={X:np.array(test_features),Y:y_cv}))
```

## Performance Evaluation

```python
plt.plot(range(len(cost_history)),cost_history, label='Training loss')
plt.plot(range(len(cost_val_history)),cost_val_history, label='Val loss')
#plt.axis([0,training_epochs,0,np.max(cost_history)])
plt.axis([0,training_epochs,0,3])
plt.legend()
_ = plt.ylim()
plt.show()
```

```python
pred_y = sess.run(y_, feed_dict={X: np.array(test_features)})
mse = tf.reduce_mean(tf.square(pred_y - y_cv))
print("MSE: %.4f" % sess.run(mse))
```
